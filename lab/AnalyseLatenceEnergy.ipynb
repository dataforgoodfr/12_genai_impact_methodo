```python
from IPython.display import display
import pandas as pd

llm_perf = pd.read_csv('data/llm_perf_filtered.csv')
ml_energy = pd.read_csv('data/aggregated_ml_energy.csv')

display(llm_perf)
display(ml_energy)
```

```python
print(llm_perf.info())
print(ml_energy.info())
```

# Initialisation

#  Calcul de la latence par token pour chaque dataset


```python
import matplotlib.pyplot as plt
import seaborn as sns

llm_perf['latency_per_token'] = llm_perf['latency'] / \
    llm_perf['response_length']
ml_energy['latency_per_token'] = ml_energy['latency'] / \
    ml_energy['response_length']

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.scatterplot(data=llm_perf, x='parameters_count',
                y='latency_per_token', hue='energy', palette='coolwarm')
plt.title('LLM Perf: Latency per Token vs Parameters Count')

plt.subplot(1, 2, 2)
sns.scatterplot(data=ml_energy, x='parameters_count',
                y='latency_per_token', hue='energy', palette='coolwarm')
plt.title('ML Energy: Latency per Token vs Parameters Count')

plt.tight_layout()
plt.show()

corr_llm = llm_perf[['latency_per_token', 'energy', 'parameters_count']].corr()
corr_ml = ml_energy[['latency_per_token', 'energy', 'parameters_count']].corr()

print("Correlation Matrix - LLM Perf:")
print(corr_llm)

print("\nCorrelation Matrix - ML Energy:")
print(corr_ml)
```

# Dataset ML Energy

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(ml_energy['latency'], ml_energy['energy'])
plt.title('Relation entre Latency et Energy')
plt.xlabel('Latency')
plt.ylabel('Energy')
plt.show()


plt.figure(figsize=(8, 6))
plt.scatter(ml_energy['response_length'], ml_energy['latency'])
plt.title('Relation entre Response Length et Latency')
plt.xlabel('Response Length')
plt.ylabel('Latency')
plt.show()
```

# Applique une régression linéaire

```python
from sklearn.linear_model import LinearRegression
import numpy as np

X = ml_energy[['parameters_count', 'response_length', 'energy']]
y = ml_energy['latency']  

model = LinearRegression()

model.fit(X, y)

latency_predictions = model.predict(X)

coefficients = model.coef_
intercept = model.intercept_

print("Coefficients:", coefficients)
print("Intercept:", intercept)

plt.figure(figsize=(10, 6))
plt.scatter(y, latency_predictions, color='blue')
plt.plot([min(y), max(y)], [min(y), max(y)], color='red')
plt.xlabel('Actual Latency')
plt.ylabel('Predicted Latency')
plt.title('Actual vs Predicted Latency')
plt.show()
```

# Clustering K-means

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

X_cluster = ml_energy[['latency', 'energy', 'parameters_count']]

scaler = StandardScaler()
X_cluster_scaled = scaler.fit_transform(X_cluster)

num_clusters = 3 

kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(X_cluster_scaled)

ml_energy['cluster_label'] = kmeans.labels_

cluster_centers = scaler.inverse_transform(kmeans.cluster_centers_)
cluster_centers_df = pd.DataFrame(cluster_centers, columns=X_cluster.columns)
print("Cluster Centers:")
print(cluster_centers_df)

plt.figure(figsize=(10, 6))
sns.scatterplot(data=ml_energy, x='latency', y='energy',
                hue='cluster_label', palette='viridis', s=100)
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1],
            marker='X', color='red', s=200, label='Cluster Centers')
plt.xlabel('Latency per token')
plt.ylabel('Total Energy')
plt.title('Clustering of Models based on Latency and Energy')
plt.legend()
plt.show()
```
