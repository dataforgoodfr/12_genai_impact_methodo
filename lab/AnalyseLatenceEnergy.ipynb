# Cas GPU A100 

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import seaborn as sns
import matplotlib.pyplot as plt

parameters_counts = [7, 12, 13, 3]

for count in parameters_counts:
    ml_energy_A100_count = ml_energy_A100[ml_energy_A100['parameters_count'] == count]

    regression_count = LinearRegression()
    X_count = ml_energy_A100_count[['response_length']]
    y_count = ml_energy_A100_count['latency']
    regression_count.fit(X_count, y_count)

    correlation_count = ml_energy_A100_count[[
        'response_length', 'latency']].corr()

    print(
        f"RÃ©gression linÃ©aire pour le dataset GPU A100 avec parameters_count={count}:")
    print("Coefficient:", regression_count.coef_)
    print("Intercept:", regression_count.intercept_)
    print("\n")
    print(
        f"Matrice de corrÃ©lation pour le dataset GPU A100 avec parameters_count={count}:")
    print(correlation_count)
    print("\n")

    sns.pairplot(ml_energy_A100_count[['response_length', 'latency']])
    plt.show()
```

# Certianes conclusions 

# En analysant les rÃ©sultats des rÃ©gressions linÃ©aires et des matrices de corrÃ©lation pour les trois cas de GPU (A40, A100, V100) avec diffÃ©rentes valeurs de parameters_count, voici quelques conclusions que l'on peut tirer :

# CorrÃ©lation entre la longueur de la rÃ©ponse et la latence :

# Dans tous les cas, il existe une corrÃ©lation positive entre la longueur de la rÃ©ponse (response_length) et la latence (latency). Cela signifie que lorsque la longueur de la rÃ©ponse augmente, la latence a tendance Ã  augmenter Ã©galement.

# Influence de la complexitÃ© du modÃ¨le (parameters_count) sur la corrÃ©lation :

# Plus la valeur de parameters_count est Ã©levÃ©e, plus la corrÃ©lation entre la longueur de la rÃ©ponse et la latence est forte. Cela suggÃ¨re que pour des modÃ¨les plus complexes, la longueur de la rÃ©ponse a une influence plus importante sur la latence.

# Comparaison entre les GPU :


# Les corrÃ©lations et les coefficients de rÃ©gression sont similaires pour les trois GPU Ã©tudiÃ©s. Cependant, les valeurs spÃ©cifiques peuvent varier lÃ©gÃ¨rement d'un GPU Ã  l'autre.

# Le GPU V100 semble montrer une corrÃ©lation lÃ©gÃ¨rement plus faible entre la longueur de la rÃ©ponse et la latence par rapport aux GPU A40 et A100, en particulier pour des modÃ¨les moins complexes (parameters_count=3).

```python
unique_gpus = llm_perf['gpu'].unique()
print("DiffÃ©rents GPU :", unique_gpus)

unique_parameters_counts = llm_perf['parameters_count'].unique()
print("DiffÃ©rents counts de paramÃ¨tres :", unique_parameters_counts)

unique_optimizations = llm_perf['optimization'].unique()
print("DiffÃ©rentes optimisations :", unique_optimizations)

unique_quantizations = llm_perf['quantization'].unique()
print("DiffÃ©rentes quantifications :", unique_quantizations)
```

```python
import pandas as pd

# DÃ©finir les intervalles pour chaque catÃ©gorie de parameters_count
intervals = [0, 10, 20, 40, float('inf')]
labels = ['petit', 'moyen', 'grand', 'trÃ¨s grand']

llm_perf['parameters_count_category'] = pd.cut(
    llm_perf['parameters_count'], bins=intervals, labels=labels, right=False)

petit_params = llm_perf[llm_perf['parameters_count_category'] == 'petit']
moyen_params = llm_perf[llm_perf['parameters_count_category'] == 'moyen']
grand_params = llm_perf[llm_perf['parameters_count_category'] == 'grand']
tres_grand_params = llm_perf[llm_perf['parameters_count_category']
                             == 'trÃ¨s grand']

print("Petit parameters_count :")
display(petit_params.head())

print("\nMoyen parameters_count :")
display(moyen_params.head())

print("\nGrand parameters_count :")
display(grand_params.head())

print("\nTrÃ¨s grand parameters_count :")
display(tres_grand_params.head())
```

```python
datasets = {
    "petit_params": petit_params,
    "moyen_params": moyen_params,
    "grand_params": grand_params,
    "tres_grand_params": tres_grand_params
}

for dataset_name, dataset in datasets.items():

    X = dataset[['parameters_count']]
    y = dataset['latency'] / dataset['response_length']

    model = LinearRegression()
    model.fit(X, y)

    print("Dataset:", dataset_name)
    print("Coefficient:", model.coef_[0])
    print("Intercept:", model.intercept_)

    plt.scatter(X, y, color='blue', label='DonnÃ©es')
    plt.plot(X, model.predict(X), color='red',
             linewidth=2, label='RÃ©gression linÃ©aire')
    plt.xlabel('DensitÃ© du modÃ¨le en token')
    plt.ylabel('Latence par token')
    plt.title(
        f'RÃ©gression linÃ©aire de la latence par token en fonction de la densitÃ© du modÃ¨le pour {dataset_name}')
    plt.legend()
    plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

X = llm_perf[['parameters_count']]
y = llm_perf['latency'] / llm_perf['response_length']

model = LinearRegression()
model.fit(X, y)

print("Coefficient:", model.coef_[0])
print("Intercept:", model.intercept_)

plt.scatter(X, y, color='blue', label='DonnÃ©es')
plt.plot(X, model.predict(X), color='red',
         linewidth=2, label='RÃ©gression linÃ©aire')
plt.xlabel('DensitÃ© du modÃ¨le en token')
plt.ylabel('Latence par token')
plt.title(
    'RÃ©gression linÃ©aire de la latence par token en fonction de la densitÃ© du modÃ¨le')
plt.legend()
plt.show()
```
