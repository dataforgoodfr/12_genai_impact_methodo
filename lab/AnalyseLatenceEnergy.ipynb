```python
# Filtrer le dataset pour chaque GPU
ml_energy_A100 = ml_energy.loc[ml_energy['gpu'] == 'A100']
ml_energy_A40 = ml_energy.loc[ml_energy['gpu'] == 'A40']
ml_energy_V100 = ml_energy.loc[ml_energy['gpu'] == 'V100']

# Afficher les datasets pour chaque GPU
print("Dataset pour le GPU A100:")
display(ml_energy_A100)
print("\n")

print("Dataset pour le GPU A40:")
display(ml_energy_A40)
print("\n")

print("Dataset pour le GPU V100:")
display(ml_energy_V100)
print("\n")
```

# Cas GPU A100 

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import seaborn as sns
import matplotlib.pyplot as plt

parameters_counts = [7, 12, 13, 3]

for count in parameters_counts:
    ml_energy_A100_count = ml_energy_A100[ml_energy_A100['parameters_count'] == count]

    regression_count = LinearRegression()
    X_count = ml_energy_A100_count[['response_length']]
    y_count = ml_energy_A100_count['latency']
    regression_count.fit(X_count, y_count)

    correlation_count = ml_energy_A100_count[[
        'response_length', 'latency']].corr()

    print(
        f"Régression linéaire pour le dataset GPU A100 avec parameters_count={count}:")
    print("Coefficient:", regression_count.coef_)
    print("Intercept:", regression_count.intercept_)
    print("\n")
    print(
        f"Matrice de corrélation pour le dataset GPU A100 avec parameters_count={count}:")
    print(correlation_count)
    print("\n")

    sns.pairplot(ml_energy_A100_count[['response_length', 'latency']])
    plt.show()
```

# Certianes conclusions 

# En analysant les résultats des régressions linéaires et des matrices de corrélation pour les trois cas de GPU (A40, A100, V100) avec différentes valeurs de parameters_count, voici quelques conclusions que l'on peut tirer :

# Corrélation entre la longueur de la réponse et la latence :

# Dans tous les cas, il existe une corrélation positive entre la longueur de la réponse (response_length) et la latence (latency). Cela signifie que lorsque la longueur de la réponse augmente, la latence a tendance à augmenter également.

# Influence de la complexité du modèle (parameters_count) sur la corrélation :

# Plus la valeur de parameters_count est élevée, plus la corrélation entre la longueur de la réponse et la latence est forte. Cela suggère que pour des modèles plus complexes, la longueur de la réponse a une influence plus importante sur la latence.

# Comparaison entre les GPU :


# Les corrélations et les coefficients de régression sont similaires pour les trois GPU étudiés. Cependant, les valeurs spécifiques peuvent varier légèrement d'un GPU à l'autre.

# Le GPU V100 semble montrer une corrélation légèrement plus faible entre la longueur de la réponse et la latence par rapport aux GPU A40 et A100, en particulier pour des modèles moins complexes (parameters_count=3).

```python
unique_gpus = llm_perf['gpu'].unique()
print("Différents GPU :", unique_gpus)

unique_parameters_counts = llm_perf['parameters_count'].unique()
print("Différents counts de paramètres :", unique_parameters_counts)

unique_optimizations = llm_perf['optimization'].unique()
print("Différentes optimisations :", unique_optimizations)

unique_quantizations = llm_perf['quantization'].unique()
print("Différentes quantifications :", unique_quantizations)
```

```python
import pandas as pd

# Définir les intervalles pour chaque catégorie de parameters_count
intervals = [0, 10, 20, 40, float('inf')]
labels = ['petit', 'moyen', 'grand', 'très grand']

llm_perf['parameters_count_category'] = pd.cut(
    llm_perf['parameters_count'], bins=intervals, labels=labels, right=False)

petit_params = llm_perf[llm_perf['parameters_count_category'] == 'petit']
moyen_params = llm_perf[llm_perf['parameters_count_category'] == 'moyen']
grand_params = llm_perf[llm_perf['parameters_count_category'] == 'grand']
tres_grand_params = llm_perf[llm_perf['parameters_count_category']
                             == 'très grand']

print("Petit parameters_count :")
display(petit_params.head())

print("\nMoyen parameters_count :")
display(moyen_params.head())

print("\nGrand parameters_count :")
display(grand_params.head())

print("\nTrès grand parameters_count :")
display(tres_grand_params.head())
```

```python
datasets = {
    "petit_params": petit_params,
    "moyen_params": moyen_params,
    "grand_params": grand_params,
    "tres_grand_params": tres_grand_params
}

for dataset_name, dataset in datasets.items():

    X = dataset[['parameters_count']]
    y = dataset['latency'] / dataset['response_length']

    model = LinearRegression()
    model.fit(X, y)

    print("Dataset:", dataset_name)
    print("Coefficient:", model.coef_[0])
    print("Intercept:", model.intercept_)

    plt.scatter(X, y, color='blue', label='Données')
    plt.plot(X, model.predict(X), color='red',
             linewidth=2, label='Régression linéaire')
    plt.xlabel('Densité du modèle en token')
    plt.ylabel('Latence par token')
    plt.title(
        f'Régression linéaire de la latence par token en fonction de la densité du modèle pour {dataset_name}')
    plt.legend()
    plt.show()
```
