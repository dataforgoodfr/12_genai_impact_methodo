```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt


llm_perf = pd.read_csv('data/llm_perf_filtered.csv')

X = llm_perf[['parameters_count']]
y = llm_perf['latency'] / llm_perf['response_length']

model = LinearRegression()
model.fit(X, y)

print("Coefficient:", model.coef_[0])
print("Intercept:", model.intercept_)

plt.scatter(X, y, color='blue', label='Donnes')
plt.plot(X, model.predict(X), color='red',
         linewidth=2, label='Rgression linaire')
plt.xlabel('Densit du modle en token')
plt.ylabel('Latence par token')
plt.title(
    'Rgression linaire de la latence par token en fonction de la densit du modle')
plt.legend()
plt.show()
```
